## Parte 1: Web Scraping con BeautifulSoup
Utilizando la biblioteca BeautifulSoup en Python, extrae información de la siguiente web https://es.wikipedia.org/wiki/Sendero_de_Gran_Recorrido. Debes
extraer la tabla de senderos de España con la siguiente información:

    1. Identificador.
    2. Denominación.
    3. Itinerario.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = "https://es.wikipedia.org/wiki/Sendero_de_Gran_Recorrido"
```

```python
def extraccion_datos(url):
    #1. Hago la llamada
    llamada=requests.get(url)
    print(f'El estado de la llamada fue {llamada.status_code}.')

    #2. Creo la sopa
    sopa_senderos=BeautifulSoup(llamada.content, 'html.parser')

#####3. Sacamos de la tabla senderos la información solicitada, guardo cada característica en una lista y, posteriormente, las junto en un diccionario

    tabla_senderos =sopa_senderos.find('table', {'class': 'wikitable'})
    un_sendero = tabla_senderos.find_all('tr')
    senderos=[]
    
    for fila in un_sendero[1:]:
        
        celdas = fila.find_all('td')
        identificador = celdas[0].text.strip()
        denominacion = celdas[1].text.strip()
        itinerario = celdas[2].text.strip()

        tupla_senderos = (identificador,denominacion,itinerario)
        senderos.append(tupla_senderos) 
    
    return senderos
```

```python
tabla_buscada = extraccion_datos(url)
df_senderos = pd.DataFrame(tabla_buscada,columns=["Identificador","Denominación","Itinerario"])

df_senderos
```
